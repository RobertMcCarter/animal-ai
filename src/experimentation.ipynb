```python
import cv2 as cv
from matplotlib import pyplot as plt

def showImage(image, title: str):
    """Show the given image with the given title in the Jupyter notebook
    """
    rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(rgb_image)
    plt.title(title)
    plt.show()
```

```python
image1 = cv.imread(r"D:\data\NRSI\__working-examples\STC_0064.JPG")
showImage(image1, 'Image 1')
```

```python
image2 = cv.imread(r"D:\data\NRSI\__working-examples\STC_0065.JPG")
showImage(image2, "Image 2")
```

```python
result = image2 - image1
showImage(result, "Difference")
```
